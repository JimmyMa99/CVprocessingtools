```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
num1 = 6101
num2 = '610h'

# 
submit1 = pd.read_csv(f'./submits/submit_{num1}.csv')
submit2 = pd.read_csv(f'./submits/submit_{num2}.csv')

# y
fig, ax = plt.subplots(1, 2, figsize=(20, 20))
sns.countplot(y='LabelName', data=submit1, ax=ax[0])
sns.countplot(y='LabelName', data=submit2, ax=ax[1])
ax[0].set_title(f'submit_{num1} score: 0.94387803360 ')
ax[1].set_title(f'submit_{num2} score: 0.94650315400 ')
ax[1].set_yticks([])
ax[1].set_ylabel('')
plt.tight_layout()
# 
for i in range(2):
    for p in ax[i].patches:
        ax[i].annotate('{:.0f}'.format(p.get_width()), (p.get_width()+10, p.get_y()+0.7))
plt.show()

# pdf
fig.savefig(f'./submit_{num1}_{num2}.pdf')
```
